#  "Data Cleaning for the Identification of Non-Paying Customers Eligible for Follow-Up Phone Outreach"

```python
import pandas as pd 
import warnings 
df=pd.read_excel("Downloads/expanded_problematic_customer_call_list.xlsx")
df
```

```python
warnings.filterwarnings("ignore")
pd.set_option("display.max_rows",None)
```

```python
df.info()
```

```python
df=df.drop(columns=["Not_Useful_Column"])
```

```python
df=df.drop_duplicates(subset=["CustomerID"])
```

```python
df=df[df["CustomerID"].astype(str).str.len()==6]
```

```python
df["CustomerID"]=df["CustomerID"].astype("int")
```

```python
df
```

```python
df["First_Name"]=df["First_Name"].fillna("Unknown").str.replace(r'[^a-zA-Z]','',regex=True)
```

```python
df["Last_Name"]=df["Last_Name"].fillna("Unknown").str.replace(r'[^a-zA-Z]','',regex=True)
```

```python
df["Last_Name"]=df["Last_Name"].str.capitalize()
```

```python
df
```

```python
df=df[(df["Phone_Number"]!="Missing") & (df["Phone_Number"]!="N/a")]
```

```python
df
```

```python
df["Phone_Number"]=df['Phone_Number'].astype(str).str.replace(r'\D', '', regex=True)
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[:3] +"-"+x[3:6]+"-"+x[6:10])
```

```python
df
```

```python
df["Paying Customer"]=df["Paying Customer"].replace("Yess","Yes")
```

```python
df
```

```python
df["Paying Customer"]=df["Paying Customer"].str.strip()
```

```python
df["Paying Customer"]=df["Paying Customer"].replace("Y","Yes")
```

```python
df
```

```python
df["Paying Customer"]=df["Paying Customer"].replace("no","No")
df["Paying Customer"]=df["Paying Customer"].replace("N","No")
df["Paying Customer"]=df["Paying Customer"].fillna('')
```

```python
df
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.strip()
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace("nO","No")
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace("n","No")
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace("yess","Yes")
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace("n0","No")
```

```python
df
```

```python
df=df[df["Do_Not_Contact"]=="No"]
```

```python
df=df[df["Paying Customer"]=="No"]
```

```python
df
```

```python
df["Address"]=df["Address"].fillna('')
```

```python
df.reset_index(drop=True,inplace=True)
```
